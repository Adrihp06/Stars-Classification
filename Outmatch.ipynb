Como los objetos que hacen match dependen del radio escogido para hacer el query en GAIA, decidi repetir dicho query aumentando el radio (0.3->0.5) para asi ver si algunos objetos no aparecen porque se han movido de más(esta puede ser una explicación)

```python
import astropy.units as u
from astropy.coordinates import SkyCoord
from astroquery.gaia import Gaia
from astropy.table import Table
import time
```

```python
#Supress default INFO logging
#This ignore the INFO message
import logging, sys
logging.disable(sys.maxsize)
import warnings
warnings.filterwarnings("ignore")
```

```python
#Take the ra and dec columns from the outliers
ra = np.array(outliers_notmatched['ra'])
dec = np.array(outliers_notmatched['dec'])
#Voy a repetir la busqueda para los objetos que no tienen match en gaia aumentando el radio de busqueda
gaia_outliers_notmatched = pd.DataFrame()
```

```python
inicio = time.perf_counter() #time counter
for i in range(0, len(ra)):
    coord = SkyCoord(ra=ra[i], dec=dec[i], unit=(u.degree, u.degree), frame='icrs')
    width = u.Quantity(5, u.arcsec)
    height = u.Quantity(5, u.arcsec)
    r = Gaia.query_object_async(coordinate=coord, width=width, height=height) #query to gaia

    gaiaquery_notmatched = r.to_pandas()
    gaia_outliers_notmatched = gaia_outliers_notmatched.append(gaiaquery_notmatched, ignore_index=True) #We save all the rows in a DataFrame variable

final = time.perf_counter()
print("Tiempo de ejecucion:", final - inicio)
```
